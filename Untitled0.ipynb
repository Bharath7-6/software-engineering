```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
n = int(input("Enter a number: "))
print("Sum of digits:", sum_of_digits(n))
```

```python
string = input("Enter numbers separated by spaces: ")
numbers = list(map(int, string.split()))
sum_even = 0
sum_odd = 0
for num in numbers:
    if num % 2 == 0:
        sum_even += num
    else:
        sum_odd += num
print("Sum of even numbers:", sum_even)
print("Sum of odd numbers:", sum_odd)
```

```python
def is_prime(digit):
  if digit < 2:
    return False
  for i in range(2, int(digit**0.5) + 1):
    if digit % i == 0:
      return False
  return True

number = input("Enter a number to find prime digits: ")
print("Prime digits in the number:")
for digit in number:
  if digit.isdigit():
    digit_int = int(digit)
    if is_prime(digit_int):
      print(digit)
  else:
    print(f"{digit} is not a digit")
```

```python
# Write a program to reverse a given number
def reverse_number(num):
    reversed_num = 0
    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num //= 10
    return reversed_num

number = int(input("Enter a number to reverse: "))
reversed_number = reverse_number(number)
print("Reversed number:", reversed_number)
```

```python
#write a program to find a palindrom of a number
def is_palindrome(num):
    original_num = num
    reversed_num = 0

    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num //= 10

    return original_num == reversed_num

number = int(input("Enter a number to check if it's a palindrome: "))
if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")
```

```python
def vacuum_world():
    goal_state = {'A': '0', 'B': '0'}

    location_input = input("Enter Location of Vacuum (A or B): ")
    status_input = input(f"Enter status of {location_input} (1 for dirty, 0 for clean): ")
    status_input_complement = input("Enter status of other tile (1 for dirty, 0 for clean): ")

    print("\nInitial Location Condition:", goal_state)

    if location_input == 'A':
        print("Vacuum is placed on Tile A")
        if status_input == '1':
            print("Tile A is Dirty.")
            print("Tile A has been Cleaned.")
        else:
            print("Tile A is already clean.")

        if status_input_complement == '1':
            print("Tile B is Dirty.")
            print("Moving RIGHT to Tile B.")
            print("Tile B has been Cleaned.")
        else:
            print("Tile B is already clean.")

    elif location_input == 'B':
        print("Vacuum is placed on Tile B")
        if status_input == '1':
            print("Tile B is Dirty.")
            print("Tile B has been Cleaned.")
        else:
            print("Tile B is already clean.")

        if status_input_complement == '1':
            print("Tile A is Dirty.")
            print("Moving LEFT to Tile A.")
            print("Tile A has been Cleaned.")
        else:
            print("Tile A is already clean.")
    else:
        print("Invalid location input. Please enter A or B.")

# Run the function
vacuum_world()
```

```python
def move(state):
    monkey_pos, monkey_status, box_pos, has_banana = state
    possible_moves = []

    if monkey_pos == 'middle' and monkey_status == 'onbox' and box_pos == 'middle' and has_banana == 'hasnot':
        possible_moves.append(('grasp', ('middle', 'onbox', 'middle', 'has')))

    if monkey_status == 'onfloor' and monkey_pos == box_pos:
        possible_moves.append(('climb', (monkey_pos, 'onbox', box_pos, has_banana)))

    for new_pos in ['door', 'window', 'middle']:
        if monkey_status == 'onfloor' and monkey_pos == box_pos and monkey_pos != new_pos:
            possible_moves.append((f'drag({monkey_pos}->{new_pos})', (new_pos, 'onfloor', new_pos, has_banana)))

    for new_pos in ['door', 'window', 'middle']:
        if monkey_status == 'onfloor' and monkey_pos != new_pos:
            possible_moves.append((f'walk({monkey_pos}->{new_pos})', (new_pos, 'onfloor', box_pos, has_banana)))

    return possible_moves

def canget(state, visited=None, depth=0):
    if visited is None:
        visited = set()

    if state[3] == 'has':
        print("Goal reached:", state)
        return True

    visited.add(state)

    for action, new_state in move(state):
        if new_state not in visited:
            print("  " * depth + f"Trying {action} -> {new_state}")
            if canget(new_state, visited, depth + 1):
                return True

    return False

def get_user_input():
    print("Define the initial state:")
    monkey_pos = input("Monkey position (door/window/middle): ").strip().lower()
    monkey_status = input("Monkey status (onfloor/onbox): ").strip().lower()
    box_pos = input("Box position (door/window/middle): ").strip().lower()
    has_banana = input("Does monkey have banana? (has/hasnot): ").strip().lower()
    return (monkey_pos, monkey_status, box_pos, has_banana)

if __name__ == "__main__":
    initial_state = get_user_input()
    result = canget(initial_state)
    print("\nResult:", "üéâ Monkey can get the banana!" if result else "‚ùå Monkey cannot get the banana.")
```

```python
def get_int_input(prompt, min_val=0):
    while True:
        try:
            val = int(input(prompt))
            if val < min_val:
                print(f"Please enter a value >= {min_val}.")
            else:
                return val
        except ValueError:
            print("Please enter a valid integer.")

def print_menu():
    print("\nList of Operations:")
    print("1. Fill Jug A completely")
    print("2. Fill Jug B completely")
    print("3. Empty Jug A completely")
    print("4. Empty Jug B completely")
    print("5. Pour from Jug A to Jug B (until B is full or A is empty)")
    print("6. Pour from Jug B to Jug A (until A is full or B is empty)")
    print("7. Pour all from Jug B to Jug A")
    print("8. Pour all from Jug A to Jug B")
    print("9. Exit")

# Initial setup with input validation
a = get_int_input("Enter Jug A Capacity: ")
b = get_int_input("Enter Jug B Capacity: ")

ai = get_int_input(f"Initially Water in Jug A (0 to {a}): ", 0)
while ai > a:
    ai = get_int_input(f"Value cannot exceed Jug A capacity ({a}): ")

bi = get_int_input(f"Initially Water in Jug B (0 to {b}): ", 0)
while bi > b:
    bi = get_int_input(f"Value cannot exceed Jug B capacity ({b}): ")

af = get_int_input(f"Final State of Jug A (0 to {a}): ", 0)
while af > a:
    af = get_int_input(f"Value cannot exceed Jug A capacity ({a}): ")

bf = get_int_input(f"Final State of Jug B (0 to {b}): ", 0)
while bf > b:
    bf = get_int_input(f"Value cannot exceed Jug B capacity ({b}): ")

steps = 0

# Process operations
while ai != af or bi != bf:
    print(f"\nCurrent State: Jug A = {ai}, Jug B = {bi}")
    print_menu()
    op = get_int_input("Enter the operation (1 - 9): ")

    if op == 1:
        ai = a
    elif op == 2:
        bi = b
    elif op == 3:
        ai = 0
    elif op == 4:
        bi = 0
    elif op == 5:
        transfer = min(ai, b - bi)
        ai -= transfer
        bi += transfer
    elif op == 6:
        transfer = min(bi, a - ai)
        bi -= transfer
        ai += transfer
    elif op == 7:
        if ai + bi > a:
            print("Cannot pour all from Jug B to Jug A (overflows).")
        else:
            ai += bi
            bi = 0
    elif op == 8:
        if ai + bi > b:
            print("Cannot pour all from Jug A to Jug B (overflows).")
        else:
            bi += ai
            ai = 0
    elif op == 9:
        break
    else:
        print("Invalid operation. Choose from 1 to 9.")
        continue

    steps += 1
    print(f"After operation {steps}: Jug A = {ai}, Jug B = {bi}")

# Final state confirmation
if ai == af and bi == bf:
    print(f"\nGoal achieved in {steps} steps! Final State: Jug A = {ai}, Jug B = {bi}")
```
